1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
Number = float (input("Enter the number to check Number id Positive Ngative or zero"))
if Number <0 :
    print('Number is negative')
elif Number>0 :
    print('number is positive')
elif Number ==0:
    print('Number is ZERO')
```

4.	Write a Python Program to Check Prime Number?

```python
CheckPrimeNumber = int(input('Enter the number'))
for i in range (2,CheckPrimeNumber):
    if CheckPrimeNumber%i==0:
        print('{0} is NOT PRIME'.format(CheckPrimeNumber))
        break
    
else:
    if CheckPrimeNumber ==1:
        print("{0} is NOT prime NOT composite".format(CheckPrimeNumber))
    else:
        print("{0} is PRIME".format(CheckPrimeNumber))
```

5.	Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
for CheckPrimeNumber in range(1,10000):
    for i in range (2,CheckPrimeNumber):
        if CheckPrimeNumber%i==0:
            break
    else:
        if CheckPrimeNumber ==1:
            print("{0} is NOT prime NOT composite".format(CheckPrimeNumber))
        else:
            print("{0} is PRIME".format(CheckPrimeNumber))
```
